```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from openpyxl.chart import LineChart, Reference
import openpyxl
```

```python
#generate artificial time series data
def generate_timeseries_df(startdate,enddate,frequency,cols,lowlimit,highlimit):
    series=pd.date_range(start=startdate,end=enddate,freq=frequency)
    df=pd.DataFrame(np.random.randint(lowlimit,highlimit,size=(len(series),cols)),index=series)

    ls=[]
    for i in range(cols):
        s="column_"
        sc=s+str(i)
        ls.append(sc)
    df.columns=ls
    return df
```

```python
#call the function
startdate="2016-01-01"
enddate="2020-12-31"
frequency="M"
cols=5
highlimit=200
lowlimit=60
df_serie=generate_timeseries_df(startdate,enddate,frequency,cols,lowlimit,highlimit)
```

```python
#create box plots
plt.figure(figsize=(16, 10))
im=df_serie.plot.box()
plt.tick_params(axis='x', labelsize=10)
plt.ylabel('y values')
#plt.savefig("boxplot.svg",format="svg",dpi=1000)
plt.savefig("boxplot.png",dpi=100)
```

```python
#set index name
df_serie.index.name="Date"
```

```python
#create time series data subplots
axs = df_serie.plot(figsize=(16, 10), subplots=True)
plt.tick_params(axis='x', labelsize=10)
#plt.savefig("subplots.png",dpi=50)
plt.savefig("subplots.svg",format="svg",dpi=1000)
plt.show()
```

```python
#generate a statistical summary and round numeric values to 2 decimals
summ=pd.DataFrame(df_serie.describe())
summ.index.name="summary"
#summ.reset_index(inplace=True)
summ.rename(columns={"index":"summary"},inplace=True)
numeric_cols = summ.iloc[:,1:].select_dtypes(include='number').columns
summ[numeric_cols] = summ[numeric_cols].round(2)
```

```python
summ
```

```python
#create excel report
with pd.ExcelWriter("excel_report.xlsx",engine="openpyxl") as writer:
    df_serie.to_excel(writer,sheet_name="time series data")
    summ.to_excel(writer,sheet_name="statistical summary")
    wb=writer.book
    ws=wb["time series data"]
    # maximum row number in the sheet
    max_row = ws.max_row
    # range of rows and cols on the sheet
    values_col0 = Reference(ws, min_col=2, min_row=1, max_col=2, max_row=max_row)
    values_col1 = Reference(ws, min_col=3, min_row=1, max_col=3, max_row=max_row)
    values_col2 = Reference(ws, min_col=4, min_row=1, max_col=4, max_row=max_row)
    values_col3 = Reference(ws, min_col=5, min_row=1, max_col=5, max_row=max_row)
    values_col4 = Reference(ws, min_col=6, min_row=1, max_col=6, max_row=max_row)
    # refer to the date
    dates = Reference(ws, min_col=1, min_row=2, max_col=1, max_row=max_row)
    # create a LineChart
    
    chart = LineChart()
    # add data of time series to the chart
    chart.add_data(values_col0, titles_from_data=True)
    chart.add_data(values_col1, titles_from_data=True)
    chart.add_data(values_col2, titles_from_data=True)
    chart.add_data(values_col3, titles_from_data=True)
    chart.add_data(values_col4, titles_from_data=True)
    # set the dates as the x axis and format it
    chart.set_categories(dates)
    chart.x_axis.number_format = 'mmm-yy'
    chart.x_axis.majorTimeUnit = 'days'
    chart.x_axis.title = 'Date'
    # add title to the chart
    chart.title = 'Time series'
    # style the chart
    s1 = chart.series[1]
    s1.graphicalProperties.line.dashStyle = 'sysDot'
    # add the chart to the cell of G12 on the sheet ws
    ws.add_chart(chart, 'G12')
    # add the boxplot to excel on cell Q12
    img = openpyxl.drawing.image.Image('boxplot.png')
    ws.add_image(img,"Q12")
```

```python
#create HTML report

df_serie_index=df_serie.copy()
df_serie_index.reset_index(inplace=True)
summ_index=summ.copy()
summ_index.reset_index(inplace=True)
page_title_text='Report'
title_text = 'Time series data'
text1 = 'This is a time serie data'
text2 = 'This is a boxplot'
time_series_text = 'Historical time series data'
summary_stats_text = 'Historical summary statistics'


# combine them together using a long f-string


html = f'''
    <html>
        <head>
            <title>{page_title_text}</title>
        </head>
        <body>
            <h1>{title_text}</h1>
            <p>{text1}</p>
            <img src='subplots.svg' width="1300">
            <p>{text2}</p>
            <img src='boxplot.svg' width="700">
            <h2>{time_series_text}</h2>
            {df_serie_index.tail(10).to_html(index=False)}
            <h2>{summary_stats_text}</h2>
            {summ_index.to_html(index=False)}
        </body>
    </html>
    '''
# 3. Write the html string as an HTML file
with open('html_report.html', 'w') as f:
    f.write(html)
```
